# CheckPoint 3: Country Analysis

```python
sorted_venture_sub_frame = venture_sub_frame.sort_values('raised_amount_usd', ascending=False) #sort venture_sub_frame in descending order
sorted_venture_sub_frame.head()
```

```python
sorted_venture_sub_frame = sorted_venture_sub_frame[sorted_venture_sub_frame['raised_amount_usd'].notnull() & sorted_venture_sub_frame['country_code'].notnull()]
country_wise_total_investment = sorted_venture_sub_frame.groupby(['country_code'])['raised_amount_usd'].agg(sum).to_dict()
sorted_country_wise_total_investment = sorted(country_wise_total_investment.items(),key=operator.itemgetter(1),reverse = True)
#operator.itemgetter(1) will pick values from dictionary that is country_wise_total_investment here
sorted_country_wise_total_investment
```

```python
#For the chosen investment type, make a data frame named top9 with the top nine countries
top9 = pd.DataFrame(data = np.array(sorted_country_wise_total_investment[0:9]),index=[1,2,3,4,5,6,7,8,9],columns=["Countries","Total investments"])
top9
```

```python
#Top 3 English-speaking countries from top9
#for getting data on english speaking countries i will scrape the web page of wikipedia
from bs4 import BeautifulSoup
import requests
import re
response = requests.get("https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language")
soup = BeautifulSoup(response.text,"html.parser")
country_tables = soup.findAll("table",class_="wikitable")
#list to store english speaking country names
english_speaking_countries=[]
for each_country_table in country_tables[0:3]:
    table_body = each_country_table.find("tbody")
    table_rows = table_body.find_all("tr")
    for each_row in table_rows[1:]:
        table_data = each_row.find_all("td")
        country = table_data[1].text.replace("\xa0","").replace("\n","")
        country = re.sub(r'\[.*]','',country)
        english_speaking_countries.append(country)# all countries with english as official language
english_speaking_countries          
```

```python
#scraping country codes for all english speaking countries        
response_two = requests.get("http://www.fao.org/countryprofiles/iso3list/en/")   
soup2 = BeautifulSoup(response_two.text,"html.parser")
country_code_table = soup2.find("table")
country_code_table_body = country_code_table.find("tbody")
country_code_table_body_rows = country_code_table_body.find_all("tr",class_="showpanel")
country_code_dict = {}
for each_row in country_code_table_body_rows:
    each_row_country_name = each_row.find("td",class_="shortname").text
    each_row_country_code = each_row.find("td",class_="iso3").text
    country_code_dict.update({each_row_country_code.lower():each_row_country_name.replace("\n","")})
country_code_dict # all countries with their codes    
```

```python
#check which countries in country_code_dict are present in english_speaking_countries_array
country_and_codes = {}
for each_country in english_speaking_countries:
    for key,val in country_code_dict.items():    
    #for each_country in english_speaking_countries:
        if re.search(r".*"+each_country.lower()+".*",val.lower()):
            country_and_codes.update({val:key})
country_and_codes
```

```python
# get top3 english speaking countries from country_and_codes dict
top_english_speaking_countries = []
for each_country in top9['Countries']:
    for key,val in country_and_codes.items():
        if each_country == val:
            top_english_speaking_countries.append(key)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
colorss = [
            '#6890F0',  # Water
            '#A040A0',  # Poison
            '#EE99AC',  # Fairy
            '#B8A038',  # Rock
            '#705898',  # Ghost
            '#98D8D8',  # Ice
            '#7038F8',  # Dragon 
            '#F08030',  # Fire
            '#6890F0',  # Water
            '#A8B820',  # Bug
            '#A8A878',  # Normal
            '#A040A0',  # Poison
            '#F8D030',  # Electric
            '#E0C068',  # Ground
]
# change the width of bars in barplot using ax.patches
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```
